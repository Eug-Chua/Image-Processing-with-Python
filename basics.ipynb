```python
# import libraries
from skimage import data, color
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

# Change Images to Grayscale

```python
# load the image
rocket = data.rocket()
rocket
```

```python
# display shape of image
rocket.shape
```

```python
# convert image to grayscale
gray_scaled_rocket = color.rgb2gray(rocket)
gray_scaled_rocket.shape
```

```python
# helper function to display images
def show_image(image, title='Image', cmap_type='gray'):
    plt.imshow(image, cmap=cmap_type)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
# show the original image
show_image(rocket, 'Original RGB image')
```

```python
# show the graysclae image
show_image(gray_scaled_rocket, 'Grayscale image')
```

# Flip Images

```python
# read the image file
image_path = 'images/01/scenery.webp'

# read image data with mpimg.imread
image = mpimg.imread(image_path)
print(f'Image shape: {image.shape}')

# flip the image
scenery_vertical_flip = np.flipud(image)

show_image(scenery_vertical_flip, 'Scenery')
```

# Histograms

Here we'll analyze the amount of red, green and blue in the image. We'll do this by computing the histogram of each of the 'rgb' channels.

```python
# read the path
image_path = 'images/01/willy-wonka.webp'

# read the image data
image = mpimg.imread(image_path)
print(f'Image shape: {image.shape}')

show_image(image, 'Willy Wonka')
```

```python
# obtain the red channel
red_channel = image[:,:, 0]

# plot the red histogram with bins in a range of 256
plt.hist(red_channel.ravel(), bins=256)

# set title and show
plt.title("Red Histogram")
plt.show()
```

```python
# obtain the green channel
green_channel = image[:,:, 1]

# plot the red histogram with bins in a range of 256
plt.hist(green_channel.ravel(), bins=256)

# set title and show
plt.title("Green Histogram")
plt.show()
```

```python
# obtain the blue channel
blue_channel = image[:,:, 2]

# plot the red histogram with bins in a range of 256
plt.hist(blue_channel.ravel(), bins=256)

# set title and show
plt.title("Blue Histogram")
plt.show()
```

With the histograms for red, green and blue, we see that the image is quite greenish.
